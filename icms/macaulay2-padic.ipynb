# Macaulay2 Padic package code demo

First, let's load the *Padic* package.

```macaulay2
needsPackage "Padic"
```

We create a new $p$-adic field "family" using `QQ_p`.

(It's a "family" as eventually we expect to have specific $p$-adic field objects in this family that use a specific precision, just like `RR_53` and `CC_100`.)

```macaulay2
QQ_7
```

$p$-adic field families are self-initializing types, i.e., we can use them to construct $p$-adic number objects.

```macaulay2
QQ_7(-1)
```

If two arguments are given, the first specifies the precision.

```macaulay2
QQ_7(10, -1)
```

$\mathbb Q_p$ is a field, so we can perform the usual field operations.

```macaulay2
x = QQ_7(-1)
y = QQ_7(1/3)
```

```macaulay2
x + y
x - y
x * y
x / y
```

Given $x\in\mathbb Q_p$, we can compute $\nu_p(x)$ and $|x|_p$.  ($\nu_p$ uses the *Valuations* package.)

```macaulay2
pVal x
abs x
```

## Example: Hensel's lemma

  Suppose $f\in\mathbb Z_p[x]$.  If there exists $\alpha_1\in\mathbb Z_p$ for which $f(\alpha_1)\equiv 0\pmod p$ and $f'(\alpha_1)\not\equiv 0\pmod p$, then there exists $\alpha\in\mathbb Z_p$ satisfying $\alpha_1\equiv\alpha\pmod p$ and $f(\alpha)=0$.

  Furthermore, $\alpha$ may be constructed by defining the recursive sequence given by Newton's method,
  \begin{equation*}
    \alpha_{n+1} = \alpha_n - \frac{f(\alpha_n)}{f'(\alpha_n)}.
  \end{equation*}
Then $f(\alpha_n)\equiv 0\pmod{p^n}$ for all $n$ and $\alpha=\lim\limits_{n\to\infty}\alpha_n$, where the limit is taken with respect to $|\cdot|_p$.

Let's find the cube root of 2 in $\mathbb Z_5$ using Hensel's lemma with $f=x^3-2$ and $f'=3x^2$.

Since $f(3)=25\equiv 0\pmod 5$ and $f'(3) = 27\not\equiv 0\pmod 5$, we begin with $\alpha_1 = 3$.

```macaulay2
newton = x -> x - (x^3 - 2)/(3*x^2);
alpha = QQ_5 3
while alpha != (alpha = newton alpha) do null
alpha
```

Let's check:

```macaulay2
alpha^3
```
